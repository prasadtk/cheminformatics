```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs

# Read SMILES strings from CSV file
csv_file_path = 'asinex.csv'
molecule_df = pd.read_csv(csv_file_path)

# Substructure SMILES string (e.g., 'CCO' for ethyl group)
substructure_smiles = 'c1cc2cccnc2s1'
substructure = Chem.MolFromSmiles(substructure_smiles)

# Search for molecules containing the substructure
matching_molecules = []
for smiles in molecule_df['SMILES']:
    molecule = Chem.MolFromSmiles(smiles)
    if molecule.HasSubstructMatch(substructure):
        matching_molecules.append(smiles)

# Convert results to a DataFrame
results_df = pd.DataFrame(matching_molecules, columns=['SMILES'])

# Save results to CSV file
output_csv_file_path = 'matching_molecules.csv'
results_df.to_csv(output_csv_file_path, index=False)

print(f'Results saved to {output_csv_file_path}')
```
