```python
from rdkit import Chem
from rdkit.Chem import Draw

# Example SMILES string
smiles = 'CCO'

# Generate a molecule object from the SMILES string
molecule = Chem.MolFromSmiles(smiles)

# Drawing options
drawing_options = Draw.MolDrawOptions()
drawing_options.bgColor = (1, 1, 1)  # background color (white)
drawing_options.dumps = True  # display atom numbers
# drawing_options.useBWAtomPalette()  # Use black and white color scheme
molecule.SetProp('_Name', 'Ethanol') # set molecule name

# Customizing atom colors
atom_colors = {}
for atom in molecule.GetAtoms():
    atom_colors[atom.GetIdx()] = (0.3, 0.2, 1)  # blueish color for atoms

# Draw the molecule
image = Draw.MolToImage(molecule, size=(300, 300), kekulize=True, wedgeBonds=True, legend="Ethanol", options=drawing_options)

# Display the image using default image viewer
image.show()
```
